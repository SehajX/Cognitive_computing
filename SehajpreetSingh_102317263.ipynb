```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Part I: Randomized Sales Data Generation
roll_no = 102317263
np.random.seed(roll_no)

# Generate sales data
sales_matrix = np.random.randint(1000, 5000, size=(12, 4))

# Define labels
months_list = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
product_categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

# Create DataFrame
sales_df = pd.DataFrame(sales_matrix, columns=product_categories, index=months_list)

# Part II: Data Manipulation and Analysis
# Display first 5 rows and summary statistics
print(sales_df.head())
print(sales_df.describe())

# Calculate total sales per category and per month
total_sales_per_category = sales_df.sum()
total_sales_per_month = sales_df.sum(axis=1)

# Calculate average sales growth between consecutive months
growth_rates = sales_df.pct_change().mean()

# Add Total Sales and Growth Rate columns
sales_df["Total Sales"] = sales_df.sum(axis=1)
sales_df["Growth Rate"] = sales_df["Total Sales"].pct_change() * 100

# Apply discount based on roll number
if roll_no % 2 == 0:
    sales_df["Electronics"] *= 0.90  # 10% discount
else:
    sales_df["Clothing"] *= 0.85  # 15% discount

# Part III: Visualizations
# Plot monthly sales trends
plt.figure(figsize=(10, 5))
for product in product_categories:
    plt.plot(sales_df.index, sales_df[product], marker='o', label=product)
plt.legend()
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# Box plot visualization
plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df[product_categories], palette="Set2")
plt.title("Sales Distribution by Category")
plt.show()
```

```python
import numpy as np

# Initialize the given array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# Element-wise absolute values
print("Element-wise Absolute Values:\n", np.abs(array))

# Compute percentiles
percentiles = {
    "Flattened": np.percentile(array, [25, 50, 75], axis=None),
    "Per Column": np.percentile(array, [25, 50, 75], axis=0),
    "Per Row": np.percentile(array, [25, 50, 75], axis=1),
}

for key, value in percentiles.items():
    print(f"\nPercentiles ({key}):\n", value)

# Compute Mean, Median, and Standard Deviation
stats = {
    "Mean": [np.mean(array), np.mean(array, axis=0), np.mean(array, axis=1)],
    "Median": [np.median(array), np.median(array, axis=0), np.median(array, axis=1)],
    "Standard Deviation": [np.std(array), np.std(array, axis=0), np.std(array, axis=1)],
}

for key, value in stats.items():
    print(f"\n{key} (Flattened, Per Column, Per Row):\n", value)
```

```python
# Function to swap two elements using a temporary variable
def swap_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

# Example
my_list = [10, 20, 30, 40, 50]
index1, index2 = 1, 3  # Swapping elements at index 1 and 3
swapped_list = swap_elements(my_list, index1, index2)

print("List after swapping:", swapped_list)
```

```python
# Function to swap two elements in a set by converting it to a list
def swap_set_elements(my_set, elem1, elem2):
    my_list = list(my_set)

    if elem1 in my_list and elem2 in my_list:
        index1, index2 = my_list.index(elem1), my_list.index(elem2)
        my_list[index1], my_list[index2] = my_list[index2], my_list[index1]

    return set(my_list)

# Example
my_set = {10, 20, 30, 40, 50}
elem1, elem2 = 20, 40  # Elements to swap
swapped_set = swap_set_elements(my_set, elem1, elem2)

print("Set after swapping:", swapped_set)
```
